```python
# text to image

import os
import sys
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw
```

```python
TELUGU_FONT = './fonts/NotoSansTelugu-Regular.ttf'
# TELUGU_FONT = './fonts/NATS.ttf'
ENGLISH_FONT = './fonts/NotoSans-Regular.ttf'
```

```python
def text2image(text,
               font_size=20,
               font_path=ENGLISH_FONT,
               image_size=(200, 50),
               image_path=None):
    if image_path is None:
        image_path = 'text2image.png'

    image = Image.new('RGB', image_size, 'white')
    draw = ImageDraw.Draw(image)
    font = ImageFont.truetype(font_path, font_size)
    draw.text((0, 0), text, (0, 0, 0), font=font)
    # image.save(image_path)

    plt.axis('off')
    plt.imshow(image)
    plt.show()


text2image('Hello, world!')
text2image('హలో, ప్రపంచం!', font_path=TELUGU_FONT)
```

```python
TELUGU_CONSONANTS = [
    'క', 'ఖ', 'గ', 'ఘ', 'ఙ', 'చ', 'ఛ', 'జ', 'ఝ', 'ఞ', 'ట', 'ఠ', 'డ', 'ఢ', 'ణ',
    'త', 'థ', 'ద', 'ధ', 'న', 'ప', 'ఫ', 'బ', 'భ', 'మ', 'య', 'ర', 'ల', 'వ', 'శ',
    'ష', 'స', 'హ', 'ళ', 'ఱ'
]
TELUGU_VOWELS = [
    'అ', 'ఆ', 'ఇ', 'ఈ', 'ఉ', 'ఊ', 'ఋ', 'ౠ', 'ఎ', 'ఏ', 'ఐ', 'ఒ', 'ఓ', 'ఔ'
]
TELUGU_VOWEL_SIGNS = [
    '', 'ా', 'ి', 'ీ', 'ు', 'ూ', 'ృ', 'ౄ', 'ె', 'ే', 'ై', 'ొ', 'ో', 'ౌ'
]
TELUGU_VOWEL_SIGNS_X_POS = ['ా', 'ి', 'ీ', 'ె', 'ే', 'ొ', 'ో', 'ౌ']
TELUGU_VOWEL_SIGNS_Y_POS = ['ృ', 'ౄ', 'ై']
TELUGU_VOWEL_SIGNS_Z_POS = ['ు', 'ూ']
TELUGU_VIRAMA = '్'
TELUGU_ANUSVARA = 'ం'
```

```python
strings = TELUGU_VOWELS + TELUGU_CONSONANTS
```

```python
# generate high quality images for all strings and save them in a folder

def generate_images(strings, folder='images'):
	if not os.path.exists(folder):
		os.makedirs(folder)

	font = ImageFont.truetype(TELUGU_FONT, 250)
	# font = ImageFont.truetype(ENGLISH_FONT, 250)
	
	for string in strings:
		image = Image.new('RGB', (1000, 500), 'white')
		draw = ImageDraw.Draw(image)
		draw.text((0, 0), string, (0, 0, 0), font=font)
		image.save(os.path.join(folder, f'image_{string}.png'))
		plt.axis('off')
		plt.imshow(image)
		plt.show()

# generate_images(strings)
generate_images("వ వ్య కి క్తి వ్యక్తి".split())
```
