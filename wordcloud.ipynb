```python
TELUGU_FONT = './fonts/NotoSansTelugu-Regular.ttf'
# TELUGU_FONT = './fonts/NATS.ttf'
ENGLISH_FONT = './fonts/NotoSans-Regular.ttf'
```

```python
center_word = "పూర్వం"
neighboring_words = generate_orthographic_neighbors2(center_word)
print(neighboring_words)

#  remove 'ుక్క'
neighboring_words = [word for word in neighboring_words if word != 'ుక్క']

print(len(neighboring_words))

freqs = {center_word: 100}
freqs.update({word: 1 for word in neighboring_words})

cloud = wordcloud.WordCloud(font_path=TELUGU_FONT,
                            width=800,
                            height=400,
                            background_color='white').generate_from_frequencies(freqs)

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
