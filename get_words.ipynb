```python
import re
TELUGU_WORD = re.compile(r'[ఀ-౿]+')
```

```python
import os
paths = os.listdir('sources')

dicts = {}

for path in paths:
	telugu_words = set(TELUGU_WORD.findall(open('sources/' + path, 'r', encoding='utf-8', errors='ignore').read()))
	print(len(telugu_words), 'telugu words in', path)

	with open('dicts/' + path + '.dict', 'w', encoding='utf-8') as f:
		for word in sorted(telugu_words):
			f.write(word + '\n')
```
